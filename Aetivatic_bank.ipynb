```python
def impute_random_nan(Data,ColName):
#Impute random value from column to NaN values for numerical features   
   
    Data[ColName+"_random"]=Data[ColName]
    random_sample = Data[ColName].dropna().sample(Data[ColName].isnull().sum(),random_state=0,replace=True)
    random_sample.index = Data[Data[ColName].isnull()].index
    Data.loc[Data[ColName].isnull(),ColName+'_random']=random_sample
```

```python
for Columns in ['inq_last_6mths','open_acc','total_acc','collections_12_mths_ex_med','acc_now_delinq','tot_coll_amt','tot_cur_bal','total_rev_hi_lim']:
    impute_random_nan(loan_data,Columns)
```

```python
def impute_nan_create_category(DataFrame,ColName):
    #Impute Nan values as "unknowm" for category features
     DataFrame[ColName] = np.where(DataFrame[ColName].isnull(),"Unknown",DataFrame[ColName])
```

```python
for Columns in ['emp_title','emp_length','batch_enrolled']:
    impute_nan_create_category(loan_data,Columns)
```

```python
 columns=['term','grade','emp_title','emp_length','sub_grade','home_ownership','pymnt_plan','purpose','title','zip_code','addr_state','initial_list_status','application_type','last_week_pay','batch_enrolled']
```

```python
for col in columns:
    #Creating codes for each category features
    loan_data[col]=loan_data[col].astype('category')
    loan_data[col] = loan_data[col].cat.codes
```

```python
from sklearn import model_selection, preprocessing, feature_selection, ensemble, linear_model, metrics, decomposition
```

```python
x_data=new_loan_data.drop(['loan_status'],axis=1)
y_data=new_loan_data[['loan_status']]
```

# Model Selection

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler
```

```python
### As data has imbalance target outputs. Its better to balance it by using sampling methods.Here I am using undersampling. 
```

```python
under = RandomUnderSampler(random_state = 2)
```

```python
X_under, y_under= under.fit_resample(x_data, y_data) 
```

```python
x_training_data, x_test_data, y_training_data, y_test_data = train_test_split(X_under, y_under, test_size = 0.3)
```

# Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(x_training_data, y_training_data)
```

```python
predictions = model.predict(x_test_data)
```

```python
from sklearn.metrics import classification_report
```

```python
report=classification_report(y_test_data, predictions)
```

```python
print(report)
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test_data, predictions))
```
